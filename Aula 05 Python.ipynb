```python
# dicionários: estrutura chave/valor
Notas = {'João': 6.0, 'Maria': 8.0, 'Pedro': 6.5}
print(Notas)
```

```python
print(Notas['João'])
```

```python
# Todas as chaves
Notas.keys()
```

```python
# Todos os valors
Notas.values()
```

```python
# Testa logicamente a existência
print('João' in Notas)
print('Fernando' in Notas)
```

```python
# Apaga um elemento
del Notas['João']
print(Notas)
```

```python
# Inclui um elemento
Notas['Ana'] = 9
print(Notas)
```

```python
# Busca um valor, se não encontrar mostra segundo parâmetro
Notas.get('Geraldo', "Não encontrado.")
```

```python
# Sets são conjuntos não ordenados de elementos não repetidos
bigdata = {'Spark', 'Hive', 'Sqop'}
print(bigdata)
```

```python
# Verifica a existência
print('Spark' in bigdata)
```

```python
# Adiciona elemento
bigdata.add('Hadoop')
bigdata
```

```python
# Número de elementos
print(len(bigdata))
```

```python
# Tuplas são listas que não podem ser alteradas
# Listas usam [], tuplas usam ()
tupla = (1,2,3,4,5,6,7)
tupla
```

```python
tupla[4]
```

```python
# Dicionário em que cada posição recebe uma tupla
dic2 = {(0, 1): 0, (1,2): 1, (3,4): 2, (4,5): 3, (6,7): 4}
dic2
```

```python
print(type(Notas))
print(type(bigdata))
print(type(tupla))
```
